```python
import pandas as pd
import numpy as np
```

```python
actuals_data = pd.read_excel('DATA/Staffing_Analysis/actuals_data_2021.xlsx')
```

```python
def load_actuals_data(path):
    actuals_data = pd.read_excel(path)
    actuals_data = actuals_data[actuals_data['Year'] == 2021]
    actuals_data = actuals_data[actuals_data['PP'].apply(lambda x: x in ('Current Modified', '01'))]
    actuals_data = actuals_data[actuals_data['Department'].apply(lambda x: x in ('Bronx Property Management', 'Brooklyn Property Management', 'Manhattan Property Management', 'Mixed Finance Conventional', 'Next Generation Operations One', 'Queens\Staten Island Property Management'))]
    actuals_data = actuals_data[actuals_data['Account_Class']=='Full-time']

    def generate_role_names(row):
        if row['CST_SUFFIX'] in ('0X', '1X'):
            return 'CARETAKER HA_X'
        elif row['CST_SUFFIX'] == '0G':
            return 'CARETAKER HA_G'
        else:
            return row['CST_NAME']

    actuals_data['ROLE'] = actuals_data.apply(lambda row: generate_role_names(row), axis=1)

    actuals_data_trimmed = actuals_data[['RC Name', 'ROLE', 'PP', 'HC']]

    actuals_chart = actuals_data_trimmed.groupby(['RC Name', 'ROLE', 'PP']).sum().unstack().reset_index()

    actuals_chart.columns = actuals_chart.columns.get_level_values(0)
    actuals_chart.columns = ['CONS', 'ROLE', 'ACTUAL', 'BUDGETED']
    
    return actuals_chart
```
